# Part A - Application Area Review (Hospitals & Medicine)

As the world collapsed at the hands of the COVID-19 pandemic, the effect on hospitals and medicine in the United Kingdom was calamitous. The disparity in surging patient numbers and shortages in medicine led to the notion that current systems were ineffective, thereby creating a consensus that change was imperative. Long before the pandemic, however, the healthcare industry was being put into question as hospital management seemed to be ineffective in resolving complex issues, furthering the exploration of artificial intelligence in this domain. Since then, many AI methods have been deployed to resolve the lack of efficiency in hospital management and medicinal supply however, their ability to work successfully is still in question.

Hospital bed allocation represents a complex issue for the NHS, one which has proven to be the catalyst of hospital disorganisation for decades. The problem becomes complex when you increase the sample size. The number of decision variables, such as, how many people might walk through the door on any given day, as well as how many beds are available become harder to predict. Moreover, the number of constraints increases such as, certain patients are only suitable for certain bed types. However, a team of data scientists as part of the Artificial Intelligence lab, 'Skunkworks' recently attempted to deploy a variety of AI techniques to resolve this problem. One of these techniques was a search strategy, specifically - Greedy search allocation. The greedy search acted so that it "allocated beds based on the best available bed at the point of admission thereby making it a fast and less resource intensive method". The method was deployed in a virtual environment over a 14-week observation period with a variety of different variables used for testing. 

Whilst the research states that the source code worked correctly and the program acted as supposed to, it does represent greater problems when distributed to a wider network. For example, the research states "the technology would need to be easily reconfigured by staff with new information about increased beds, changed ward layouts or flu admission peaks" which represents a logistical issue of maintenance. If this system was to be deployed to all NHS hospitals throughout the country, each system would have to have the ability to be quickly changed depending on such constraints as mentioned above. Therefore staff would have to be trained in doing so or they would need to employ a person(s) who would be able to monitor changes and change the source code quickly and accordingly. Another problem that can be observed is that a greedy search has no heuristics, it is only choosing the bed based on the quickest allocation period, therefore it will fail in larger environments where searches become more complex. For example, a greedy search would be unable to interpret information such as different floors requiring longer routes because of stairs/elevators, it only interprets 'G', the distance, not any other information such as 'H', heuristics, like in an A* search.

At the same time bed allocation was leading to disarray in hospitals up and down the country, the availability of medicine and machinery was also causing issues, more specifically, ventilation machines. An excess demand of these machines throughout the pandemic put immense strain on the NHS, and trying to allocate the machines in the fairest and most efficient way became an enormous task causing frequent disruption amongst patients and staff. As the pandemic brought about an unprecedented amount of hospital patients the NHS had no systems prepared to resolve the issues mentioned above and therefore they relied on human decision making to allocate ventilators. Many had questioned if new computer systems could make ventillator allocation quicker and fairer in order to reduce the strain on what soon became an invaluable necessity.

Research from (de Godoy et al. 2023) argues that the allocation of ventilators should not be a solution-based problem but rather a prevention problem. Their research is based on the idea that if hospitals became better at predicting who would need a ventilator rather than waiting until the point when one is essential, they could better allocate them due to greater planning/preparation time. Based off this idea, the research deploys a convolutional neural network (CNN) aiming to highlight topological features within a CT scan where signs that use of a ventilator are imminent.

The image above represents how the CNN was tested, trained, and then completed. In comparison to traditional CNN's the research group decided to use K-fold cross-validation in which, a different fold is used in every new training phase for validation in order to reduce bias and errors in the training phase. Overall, the research which "consisted of chest CT examinations from 823 patients with COVID-19" represented that the CNN was somewhat effective in predicting which patients would need a ventilator and as the research states "makes this approach ideal for identifying high-risk patients and predicting the minimum number of ventilators and critical care beds that will be required". The research also represented a specificity score of 92%, however an overall accuracy of only 56% which is relatively low but would still be considered an improvement on current methods. The K-fold cross-validation was evidently successful in reducing errors in the CNN however it would not change the original accuracy measures. It is evident that the notion of prevention rather than solution is helpful but deciding which AI technique is best for this is a monumental task.

```python
import pandas as pd
#Helps when analysing data
from sklearn.preprocessing import MinMaxScaler 
#MinMaxScaler function allows the data be normalised on a scale of 0-1
import numpy as np 
#Will help processing the data lists towards the end
from sklearn.cluster import KMeans 
#Allows for the clusteroids to be completed faster and easier 
import matplotlib.pyplot as plt 
#Allows data to be visualised 


file_path = '/Users/maxbenson/Desktop/MedicalData.csv'
Medicaldf = pd.read_csv(file_path)
#Searching for dataset on computer and then reading it 

print(Medicaldf)
```

Now that the dataframe has been uploaded, it must be normalised. In order to do this I will use the MinMaxScaler function as previously seen.

```python
numeric_columns = Medicaldf.select_dtypes(include=['number']).columns #Selecting only the numeric columns

scaler = MinMaxScaler() #intialising the function
Medicaldf[numeric_columns] = scaler.fit_transform(Medicaldf[numeric_columns]) #applying the function to the numeric columns

print(Medicaldf)
```

As we can now see, all numeric data has been normalised using the MinMaxScaler function so that the data is scaled between 1 and 0. Now that this has been complete, the next necessary step is the clustering. The desire is to create 4 clusters however I will attempt to use the elbow algorithm to determine the accuracy of using 4. If it seems unsuitable then I will have to ammend the number of groups at the end.

```python
numeric_columns = Medicaldf.select_dtypes(include=['number']) #Again, only using numeric columns

k_values = range(1,6) #Not interested in anything over K=6 
inertia_values =[]

for k in k_values:
    kmeans=KMeans(n_clusters=k, random_state=30)
    kmeans.fit(numeric_columns)
    inertia_values.append(kmeans.inertia_)
#Loop iterates over each value of K, applying Kmeans to each and appending the inertia each time

plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('inertia')
plt.title('Elbow Method')
plt.show()
#plots the graph 
```

Now we have created the elbow graph, it is important to determine if k=4 is an appropriate number of clusters. In order to determine this, observing the graph I need to locate the point at which inertia starts to decrease in a linear degree. This initially happens at k=2 however it also happens again at k=4, therefore I will continue the prototype under the premise k=4.

Now that the number of K's has been established, I can implement the clustering algorithm to visualise the data in k=4, as shown below.
